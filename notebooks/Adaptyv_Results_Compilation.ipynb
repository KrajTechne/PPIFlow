```python
import pandas as pd
import os
fmwk_ids = ['7eow', '8coh']
path_mpnns = [f'/Volumes/sandbox/denovotrial/ppiflow/Mar25_FMWK{fmwk_id}_RBX1_7_bbones/round2_mpnn/boltz_predictions/round2_mpnn_boltz.csv' for fmwk_id in fmwk_ids]
path_mpnns.append('/Volumes/sandbox/denovotrial/ppiflow/Mar24_FMWK710_RBX1_7_bbones/round2_mpnn/boltz_predictions/round2_mpnn_boltz.csv')

df_mpnns = pd.concat([pd.read_csv(path) for path in path_mpnns])
print("Number of Predicted Structures: ", df_mpnns.shape[0])
df_mpnns.head()
```

```python
df_designs_aggregated = df_mpnns.groupby('design_name').agg({'iptm' : 'mean'}).reset_index()
df_100_designs = df_designs_aggregated[df_designs_aggregated['iptm'] > 0.72]
designs_100 = df_100_designs['design_name'].tolist()
df_designs_aggregated.sort_values(by = 'iptm', ascending= False).head(100)
```

```python
from StrucTools import *
import biotite.structure as struc
import biotite.sequence as seq
df_top_designs_full_metrics = df_mpnns[df_mpnns['design_name'].isin(designs_100)]
top_designs_paths = df_top_designs_full_metrics['path_structure']
top_designs_seqs = []
for index, row in df_top_designs_full_metrics.iterrows():
    path_structure = row['path_structure']
    design_name = row['design_name']
    atom_array_complex = extract_atom_array(path_structure)
    seqs, chain_starts = struc.to_sequence(atom_array_complex)
    seq_binder, seq_target = str(seqs[0]), str(seqs[1])
    top_designs_seqs.append({'design_name': design_name, 'seq_binder': seq_binder, 'target_chain': seq_target})

df_designs_seqs = pd.DataFrame(top_designs_seqs)
df_designs_seqs.head()
```

```python
df_top_designs_final = df_designs_seqs.drop_duplicates(subset='design_name')
print("Number of Top Final Designs: ", df_top_designs_final.shape[0])
#df_top_designs_final.to_csv("/Volumes/sandbox/denovotrial/ppiflow/Adaptyv_RBX1_Top_Designs_Seqs.csv")
df_top_designs_final.head()
```
